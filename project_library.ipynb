```python
# 코드 장성 전 필요한 module 꼭 가져기기
import pandas as pd
```

```python
# 1번
# 조회하려는 엑셀 파일 불러오기
df_excel = pd.read_excel('crimes.xlsx')
# 출력되는 내용이 길어지므로 간단히 조회
df_excel.head()
```

```python
# 2번
office_to_district = {'서대문서': '서대문구', '수서서': '강남구', '강서서': '강서구', '서초서': '서초구',
'서부서': '은평구', '중부서': '중구', '종로서': '종로구', '남대문서': '중구',
'혜화서': '종로구', '용산서': '용산구', '성북서': '성북구', '동대문서': '동대문구',
'마포서': '마포구', '영등포서': '영등포구', '성동서': '성동구', '동작서': '동작구',
'광진서': '광진구', '강북서': '강북구', '금천서': '금천구', '중랑서': '중랑구',
'강남서': '강남구', '관악서': '관악구', '강동서': '강동구', '종암서': '성북구', 
'구로서': '구로구', '양천서': '양천구', '송파서': '송파구', '노원서': '노원구', 
'방배서': '서초구', '은평서': '은평구', '도봉서': '도봉구'}
# '관서명'을 인덱스 기준으로 새로 매핑하고 '구별'이라는 새 컬럼 추가
df_excel['구별'] = df_excel['관서명'].map(office_to_district)
# 논 값을 '구 없음'으로 채우기
df_excel_filled = df_excel.fillna('구 없음')
df_excel_filled.head()
```

```python
# 3번
# '구별'을 인덱스 기준으로 그룹하고 같은 그룹은 aggfunc로 sum한다
df_excel_pvt = pd.pivot_table(df_excel_filled, index = '구별', aggfunc = 'sum')
# 불필요한 컬럼인 '관서명'을 제거한다
df_excel_pivot = df_excel_pvt.drop(columns = '관서명')
df_excel_pivot.head(8)
```

```python
# 4번
# '구 없음' 열을 제거해준다
df_excel_drop = df_excel_pivot.drop(labels = '구 없음')
df_excel_drop.head(7)
```

```python
# 컬럼 이름의 공통된 단어를 리스트로 설정한 뒤 for문과 format으로 한 번에 계산해주기
for crime in ['강간', '강도', '살인', '절도', '폭력']
    df_excel_drop[f'{crime}검거율'] =  (df_excel_drop[f'{crime}(검거)'] / df_excel_drop[f'{crime}(발생)']) * 100
# 위의 format과는 살짝 달라 따로 해주기
df_excel_drop['검거율'] = (df_excel_drop['소계(검거)'] / df_excel_drop['소계(발생)']) * 100
```

```python
# 6번
# 원하는 컬럼만 제거하기
to_del = ['강간(검거)', '강도(검거)', '살인(검거)', '소계(검거)', '소계(발생)', '절도(검거)', '폭력(검거)']
for col in to_del:
    del df_excel_drop[col]
```

```python
# 7번
# 남은 컬럼 이름 재설정 해주기
to_rename = {'강간(발생)': '강간',  '강도(발생)': '강도', '살인(발생)': '살인', '절도(발생)': '절도', '폭력(발생)': '폭력'}
df_excel_drop.rename(columns = to_rename, inplace = False)
```
